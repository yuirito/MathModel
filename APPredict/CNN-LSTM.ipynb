```python
import torch.nn as nn
import torch
```

```python
class ConvNet(nn.Module):
    def __init__(self):
        super(ConvNet, self).__init__()
        self.conv= nn.Conv2d(in_channels=72, out_channels=72, kernel_size=(4,1), stride=1)
        
        
    def forward(self, x):
        out = self.conv(x)
        
        
        return out
    
 
    
     
    
```
