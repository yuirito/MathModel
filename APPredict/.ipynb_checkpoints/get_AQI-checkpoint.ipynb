```python
import math

def max_index(lst_int):
    index = []
    max_n = max(lst_int)
    for i in range(len(lst_int)):
        if lst_int[i] == max_n:
            index.append(i)
    return index
def getAQI(list):
    IAQI_temp=[]
    IAQI=[
    [0,50,100,150,200,300,400,500],
    [0,50,150,475,800,1600,2100,2620],
    [0,40,80,180,280,565,750,940],
    [0,50,150,250,350,420,500,600],
    [0,35,75,115,150,250,350,500],
    [0,100,160,215,265,800],
    [0,2,4,14,24,36,48,60]]
    for i in range(len(list)):
        Cp=list[i]
        IAQI_poll=IAQI[i+1]
        BP_Hi,BP_Lo=0,0
        IAQI_Hi,IAQI_Lo=0,0
        for j in range(len(IAQI_poll)):
            if  Cp<IAQI_poll[j]:
                BP_Hi=IAQI_poll[j]
                IAQI_Hi=IAQI[0][j]
                if j==0:
                    BP_Lo=0
                    IAQI_Lo=0
                else:
                    BP_Lo=IAQI_poll[j-1]
                    IAQI_Lo=IAQI[0][j-1]
                break
        if BP_Hi==BP_Lo:
            IAQI_t=0
        else:
            IAQI_t=math.ceil(((IAQI_Hi-IAQI_Lo)/(BP_Hi-BP_Lo))*(Cp-BP_Lo)+IAQI_Lo)
        IAQI_temp.append(IAQI_t)
    return max(IAQI_temp),max_index(IAQI_temp)
```

```python
pn=0
AQI_loss=0
for i in range(len(act)):
    list_pre=pre.loc[i,:]
    list_act=act.loc[i,:]
    ap,mp=getAQI(list_pre)
    aa,ma=getAQI(list_act)
    AQI_loss=AQI_loss+(ap-aa)**2
    if set(ma)<=set(mp):
        pn=pn+1
pn=pn/len(act)
AQI_loss =  AQI_loss/len(act)
```
