```python
df_pre.index = df_pre.index.astype('datetime64[ns]')
df_pre.index
```

```python
df_real_trn.index
```

```python
id = []
for index in range(0, len(df_real_trn.index)):
    if df_real_trn.index[index] not in df_pre.index:
        print(f'Index {df_real_trn.index[index]} is not in df_pre.index')
        id.append(index)

df_real_trn.drop(index=df_real_trn.index[id], inplace=True)
df_real_trn
```

```python
df_real_trn.info()
```

```python
df_real_trn.fillna(0, inplace=True)
```

```python
df_pre_np = df_pre.to_numpy()
df_real_trn_np = df_real_trn.to_numpy()
```

```python
mse = ((df_pre_np - df_real_trn_np)**2).mean(axis=0)
```

```python
mse.reshape(-1,6)
```

```python
df_eval = pd.DataFrame(data = mse.T, index=df_pre.columns, columns=['MSE'])
```

```python
df_eval.to_csv('模型评估.csv')
df_real_trn.to_csv('监测点A实测预处理.csv')
```
