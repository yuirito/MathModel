# 计算AQI

```python
import math

def max_index(lst_int):
    index = []
    max_n = max(lst_int)
    for i in range(len(lst_int)):
        if lst_int[i] == max_n:
            index.append(i)
    return index
def getAQI(list):
    IAQI_temp=[]
    IAQI=[
    [0,50,100,150,200,300,400,500],
    [0,50,150,475,800,1600,2100,2620],
    [0,40,80,180,280,565,750,940],
    [0,50,150,250,350,420,500,600],
    [0,35,75,115,150,250,350,500],
    [0,100,160,215,265,800],
    [0,2,4,14,24,36,48,60]]
    for i in range(len(list)):
        Cp=list[i]
        IAQI_poll=IAQI[i+1]
        BP_Hi,BP_Lo=0,0
        IAQI_Hi,IAQI_Lo=0,0
        for j in range(len(IAQI_poll)):
            if  Cp<IAQI_poll[j]:
                BP_Hi=IAQI_poll[j]
                IAQI_Hi=IAQI[0][j]
                if j==0:
                    BP_Lo=0
                    IAQI_Lo=0
                else:
                    BP_Lo=IAQI_poll[j-1]
                    IAQI_Lo=IAQI[0][j-1]
                break
        if BP_Hi==BP_Lo:
            IAQI_t=0
        else:
            IAQI_t=math.ceil(((IAQI_Hi-IAQI_Lo)/(BP_Hi-BP_Lo))*(Cp-BP_Lo)+IAQI_Lo)
        IAQI_temp.append(IAQI_t)
    return max(IAQI_temp),max_index(IAQI_temp)
```

```python
AQI = []
AQII = []
for i in range(0,len(df_A_avg)):
    pol = df_A_avg.iloc[i,15:21].to_list()
    q, qq = getAQI(pol)
    AQI.append(q)
    AQII.append(qq)
AQI
```

```python
df_A_avg['AQI']=AQI
df_A_avg['MainP'] = AQII
```
